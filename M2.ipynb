## DATA VISUALIZATION AND DATA ANALYSIS

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('RELIANCE.csv')
df.head()
```

```python
df=df[["Open","High","Low","Volume","Close"]]
```

```python
df.columns
```

```python
df = df.dropna()
```

```python
df=df.drop_duplicates()
```

```python
df.tail()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(df['Open'],color='orange')

plt.subplot(1,2,2)
plt.hist(df['Close'],color='tomato')
```

```python
df.hist(figsize=(15,55),layout=(15,4), color='gold')
plt.show()
```

```python
df['Volume'].hist(figsize=(10,5),color='brown')
```
